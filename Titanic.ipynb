```python
df=pd.read_csv('/kaggle/input/titanic/train.csv')
df.head()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df.head()
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.Embarked.fillna('S',inplace=True)
df.Embarked.isna().sum()
```

```python
df.Sex=df.Sex.map({'male':1,'female':0})
```

```python
df['Sex'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
```

```python
df2=ohe.fit_transform(df[['Embarked']]).toarray()
```

```python
dfe=pd.DataFrame(df2,columns=ohe.get_feature_names_out())
dfe
```

```python
df=pd.concat([df,dfe],axis=1)
```

```python
df.drop('Embarked',axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x=df.drop('Survived',axis=1)
y=df['Survived']
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(x,y)
```

```python
testdf=pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
ids=testdf.PassengerId
ids
```

```python
testdf.head()
```

```python
testdf['Age'].fillna(df['Age'].mean(),inplace=True)
testdf.Embarked.fillna('S',inplace=True)
testdf.Fare.fillna(testdf.Fare.mean(),inplace=True)
```

```python
testdf.isnull().any()
```

```python
testdf.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
testdf.Sex=testdf.Sex.map({'male':1,'female':0})
df3=ohe.transform(testdf[['Embarked']]).toarray()
dfe=pd.DataFrame(df3,columns=ohe.get_feature_names_out())
testdf = pd.concat([testdf.drop('Embarked', axis=1), dfe], axis=1)
```

```python
ypred=dt.predict(testdf)
```

```python
submission=pd.DataFrame({
    'PassengerId': ids,
    'Survived': ypred
})
submission.to_csv('submission.csv',index=False)
```
