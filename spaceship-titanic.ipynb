```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df.drop(['PassengerId','Name'],axis=1,inplace=True)
df.head()
```

```python
df.HomePlanet.value_counts()
```

```python
df.HomePlanet.fillna('Earth',inplace=True)
df.HomePlanet.isnull().sum()
```

```python
df.Cabin.value_counts()
```

```python
df.VIP.value_counts()
```

```python
df.VIP.isnull().sum()
```

```python
df.VIP.fillna(False,inplace=True)
```

```python
df.VIP.isnull().sum()
```

```python
df.Destination.value_counts()
```

```python
df.Destination.isnull().sum()
```

```python
df.Destination.fillna('TRAPPIST-1e',inplace=True)
df.Destination.value_counts()
```

```python
df.drop(['Cabin'],axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
ohe1=OneHotEncoder()
ohe2=OneHotEncoder()
le=LabelEncoder()
```

```python
df.VIP=le.fit_transform(df.VIP)
df.VIP.value_counts()
```

```python
df1=ohe1.fit_transform(df[['Destination']])
df2=ohe2.fit_transform(df[['HomePlanet']])
```

```python
df1=pd.DataFrame(df1.toarray(),columns=ohe1.get_feature_names_out())
df2=pd.DataFrame(df2.toarray(),columns=ohe2.get_feature_names_out())
df=pd.concat([df,df1,df2],axis=1)
df.head()
```

```python
df.drop(['HomePlanet','Destination'],axis=1,inplace=True)
df.head()
```

```python
le2=LabelEncoder()
le3=LabelEncoder()
df.CryoSleep=le2.fit_transform(df.CryoSleep)
df.Transported=le3.fit_transform(df.Transported)
df.head()
```

```python
df.VRDeck=df.VRDeck.fillna(df.VRDeck.mean())
df.Spa=df.Spa.fillna(df.Spa.mean())
df.ShoppingMall=df.ShoppingMall.fillna(df.ShoppingMall.mean())
df.FoodCourt=df.FoodCourt.fillna(df.FoodCourt.mean())
df.Age=df.Age.fillna(df.Age.mean())
df.RoomService=df.RoomService.fillna(df.RoomService.mean())
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
params={
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto'],
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid']
}
grid=GridSearchCV(SVC() , param_grid=params, cv=2, scoring='accuracy',verbose=3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=df.drop('Transported',axis=1)
y=df['Transported']
x=scaler.fit_transform(x)
```

```python
grid.fit(x,y)
```

```python
grid.best_params_, grid.best_score_
```

```python
testdf=pd.read_csv('test.csv')
passenger_ids = testdf['PassengerId'] # Save PassengerId before dropping
testdf.drop(['PassengerId','Name','Cabin'],axis=1,inplace=True)
testdf['HomePlanet'] = testdf['HomePlanet'].fillna('Earth')
testdf['VIP'] = testdf['VIP'].fillna(False)
testdf['Destination'] = testdf['Destination'].fillna('TRAPPIST-1e')
testdf.VIP=le.fit_transform(testdf.VIP)
testdf1=ohe1.transform(testdf[['Destination']]).toarray() # Convert to dense array
testdf2=ohe2.transform(testdf[['HomePlanet']]).toarray() # Convert to dense array
testdf1=pd.DataFrame(testdf1,columns=ohe1.get_feature_names_out())
testdf2=pd.DataFrame(testdf2,columns=ohe2.get_feature_names_out())
testdf=pd.concat([testdf,testdf1,testdf2],axis=1)
testdf.drop(['HomePlanet','Destination'],axis=1,inplace=True)
testdf.CryoSleep=le2.transform(testdf.CryoSleep)
# testdf.Transported=le3.transform(testdf.Transported) # Do not transform the target variable in the test set
testdf['VRDeck'] = testdf['VRDeck'].fillna(testdf['VRDeck'].mean())
testdf['Spa'] = testdf['Spa'].fillna(testdf['Spa'].mean())
testdf['ShoppingMall'] = testdf['ShoppingMall'].fillna(testdf['ShoppingMall'].mean())
testdf['FoodCourt'] = testdf['FoodCourt'].fillna(testdf['FoodCourt'].mean())
testdf['Age'] = testdf['Age'].fillna(testdf['Age'].mean())
testdf['RoomService'] = testdf['RoomService'].fillna(testdf['RoomService'].mean())

# Ensure the test data has the same columns as the training data before scaling
# The 'Transported' column was dropped from the training data before scaling
# Although 'Transported' doesn't exist in the test set, we need to align columns
# This can be done by re-indexing the test dataframe with the training columns
testdf = testdf.reindex(columns=df.drop('Transported', axis=1).columns, fill_value=0)

testdf = scaler.transform(testdf) # Scale the test data

svc=SVC(C= 1, gamma= 'auto', kernel= 'rbf')
svc.fit(x,y)
pred=svc.predict(testdf)
preds = pd.Series(pred).map({0: False, 1: True})
submission = pd.DataFrame({
    'PassengerId': passenger_ids,  # Use PassengerId from the test set
    'Transported': preds
})

submission.to_csv('submission.csv', index=False)
```
